```python
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox
import json


class Pelanggan:
    def __init__(self):
        self.root = root
        with open("data.json", "r") as j:
            self.data_pelanggan = json.load(j)

    def clear(self):
        for item in self.root.winfo_children():
            item.destroy()

    def pelanggan(self):
        self.clear()
        self.judul = Label(self.root, text="Pelanggan")
        self.judul.grid(row=0, column=0, padx=10, pady=5, sticky=W)
        self.btn_tambah = Button(
            self.root, text="Tambah Pesanan", command=self.tambah_pesanan
        )
        self.btn_tambah.grid(row=1, column=0, padx=10, pady=5, sticky=W)

    def tambah_pesanan(self):
        self.clear()
        self.temp_data = {
            "roti": "",
            "rasa": "",
            "topping": "",
        }
        self.judul = Label(self.root, text="Tambah Pesanan")
        self.judul.grid(row=0, column=0, padx=10, pady=5, sticky=W)
        self.label_nama = Label(self.root, text="Nama : ")
        self.label_nama.grid(row=1, column=0, padx=10, pady=5, sticky=W)
        self.entry_nama = Entry(self.root)
        self.entry_nama.grid(row=1, column=1, padx=10, pady=5, sticky=W)
        self.label_roti = Label(self.root, text="Roti : ")
        self.label_roti.grid(row=2, column=0, padx=10, pady=5, sticky=W)

        roti = ["", "Pastry", "Polos"]
        clicked_roti = StringVar()
        clicked_roti.set("Pilih Roti")
        self.entry_roti = OptionMenu(
            self.root, clicked_roti, *roti, command=self.get_roti
        )
        self.entry_roti.grid(row=2, column=1, padx=10, pady=5, sticky=W)

        self.label_rasa = Label(self.root, text="Rasa : ")
        self.label_rasa.grid(row=3, column=0, padx=10, pady=5, sticky=W)
        rasa = ["", "Coklat", "Keju", "Strawberry", "Blueberry", "Tiramisu", "Matcha"]
        clicked_rasa = StringVar()
        clicked_rasa.set("Pilih Rasa")
        self.entry_rasa = OptionMenu(
            self.root, clicked_rasa, *rasa, command=self.get_rasa
        )
        self.entry_rasa.grid(row=3, column=1, padx=10, pady=5, sticky=W)
        self.label_topping = Label(self.root, text="Topping : ")
        self.label_topping.grid(row=4, column=0, padx=10, pady=5, sticky=W)
        topping = ["", "Kacang", "Oreo", "Keju"]
        clicked_topping = StringVar()
        clicked_topping.set("Pilih Topping")
        self.entry_topping = OptionMenu(
            self.root, clicked_topping, *topping, command=self.get_topping
        )
        self.entry_topping.grid(row=4, column=1, padx=10, pady=5, sticky=W)
        self.label_jumlah = Label(self.root, text="Jumlah : ")
        self.label_jumlah.grid(row=5, column=0, padx=10, pady=5, sticky=W)
        self.entry_jumlah = Entry(self.root)
        self.entry_jumlah.grid(row=5, column=1, padx=10, pady=5, sticky=W)
        self.btn_tambah_pesanan = Button(self.root, text="Tambah", command=self.tambah)
        self.btn_tambah_pesanan.grid(row=6, column=0, padx=10, pady=5, sticky=W)
        self.btn_kembali = Button(self.root, text="Kembali", command=self.pelanggan)
        self.btn_kembali.grid(row=6, column=1, padx=10, pady=5, sticky=W)

    def get_roti(self, value):
        self.temp_data["roti"] = value

    def get_rasa(self, value):
        self.temp_data["rasa"] = value

    def get_topping(self, value):
        self.temp_data["topping"] = value

    def tambah(self):
        self.label_kon_judul = Label(self.root, text="Pesanan Ditambahkan")
        self.label_kon_judul.grid(row=7, column=0, padx=10, pady=5, sticky=W)
        self.label_kon_nama = Label(self.root, text="Nama : " + self.entry_nama.get())
        self.label_kon_nama.grid(row=8, column=0, padx=10, pady=5, sticky=W)
        self.label_kon_jumlah = Label(
            self.root, text="Jumlah : " + self.entry_jumlah.get()
        )
        self.label_kon_jumlah.grid(row=9, column=0, padx=10, pady=5, sticky=W)
        self.data_pelanggan.append(
            {
                "nama": self.entry_nama.get(),
                "roti": self.temp_data["roti"],
                "rasa": self.temp_data["rasa"],
                "topping": self.temp_data["topping"],
                "jumlah": self.entry_jumlah.get(),
                "pembayaran": False,
            }
        )
        with open("data.json", "w") as j:
            json.dump(self.data_pelanggan, j, indent=4)
        print(self.data_pelanggan)
        messagebox.showinfo("Berhasil", "Pesanan Berhasil ditambahkan !!!")
        self.pelanggan()


class Kasir:
    def __init__(self):
        self.root = root
        with open("data.json", "r") as j:
            self.data_pelanggan = json.load(j)

    def clear(self):
        for item in self.root.winfo_children():
            item.destroy()

    def kasir(self):
        self.clear()
        self.judul = Label(self.root, text="Kasir")
        self.judul.grid(row=0, column=0, padx=10, pady=5, columnspan=3)
        self.password = Label(self.root, text="Masukan Password : ")
        self.password.grid(row=2, column=0, padx=10, pady=5, sticky=W)
        self.entry_password = Entry(self.root)
        self.entry_password.grid(row=2, column=1, padx=10, pady=5, sticky=W)
        self.btn_login = Button(self.root, text="Login", command=self.login)
        self.btn_login.grid(row=2, column=2, padx=10, pady=5, sticky=W)

    def login(self, passwd=""):
        if self.entry_password.get() == "ve" or passwd == "ve":
            self.clear()
            self.judul = Label(self.root, text="Selamat Datang Kasir")
            self.judul.grid(row=0, column=0, padx=10, pady=5, sticky=W)
            self.btn_tampil = Button(
                self.root, text="Tampilkan Pesanan", command=self.tampil
            )
            self.btn_tampil.grid(row=2, column=0, padx=10, pady=5, sticky=W)
            self.btn_selesai = Button(
                self.root, text="Selesaikan Pesanan", command=self.selesai
            )
            self.btn_selesai.grid(row=2, column=1, padx=10, pady=5, sticky=W)

        else:
            messagebox.showerror("Password Salah", "Masukan Password Dengan Benar")
            self.entry_password.delete(0, END)

    def selesai(self):
        self.clear()
        self.judul = Label(self.root, text="Daftar Pesanan")
        self.judul.grid(row=0, column=0, padx=10, pady=5, columnspan=3)
        i = 1
        table_kolom = ("no", "nama", "roti", "rasa", "topping", "jumlah", "pembayaran")
        self.table = Treeview(
            self.root,
            columns=table_kolom,
        )
        self.table.column("#0", stretch=NO, width=0)
        for j in table_kolom:
            self.table.column(j, stretch=YES, anchor=CENTER)
            self.table.heading(j, text=j, anchor=CENTER)
        for j in self.data_pelanggan:
            if j["pembayaran"]:
                self.table.insert(
                    "",
                    "end",
                    iid=i,
                    values=(
                        i,
                        j["nama"],
                        j["roti"],
                        j["rasa"],
                        j["topping"] if j["topping"] != "" else "-",
                        j["jumlah"],
                        j["pembayaran"],
                    ),
                )
                i += 1
        self.table.grid(row=1, column=0, padx=10, pady=5, columnspan=3)
        self.table.bind("<Double-1>", self.hapus)
        self.btn_kembali = Button(self.root, text="Kembali", command=self.kasir)
        self.btn_kembali.grid(row=2, column=0, padx=10, pady=5)

    def hapus(self, event):
        temp = self.table.item(self.table.selection())["values"]
        for j in self.data_pelanggan:
            if j["nama"] == temp[1]:
                self.data_pelanggan.remove(j)
                with open("data.json", "w") as k:
                    json.dump(self.data_pelanggan, k, indent=4)
                self.selesai()

    def tampil(self):
        self.clear()
        self.judul = Label(self.root, text="Daftar Pesanan")
        self.judul.grid(row=0, column=0, padx=10, pady=5, columnspan=3)
        i = 1
        table_kolom = ("no", "nama", "roti", "rasa", "topping", "jumlah", "pembayaran")
        self.table = Treeview(
            self.root,
            columns=table_kolom,
        )
        self.table.column("#0", stretch=NO, width=0)
        for j in table_kolom:
            self.table.column(j, stretch=YES, anchor=CENTER)
            self.table.heading(j, text=j, anchor=CENTER)
        for j in self.data_pelanggan:
            self.table.insert(
                "",
                "end",
                iid=i,
                values=(
                    i,
                    j["nama"],
                    j["roti"],
                    j["rasa"],
                    j["topping"] if j["topping"] != "" else "-",
                    j["jumlah"],
                    j["pembayaran"],
                ),
            )
            i += 1
        self.table.grid(row=1, column=0, padx=10, pady=5, columnspan=3)
        self.table.bind("<Double-1>", self.payment)
        self.btn_kembali = Button(self.root, text="Kembali", command=self.kasir)
        self.btn_kembali.grid(row=2, column=0, padx=10, pady=5)

    def payment(self, event):
        temp = self.table.item(self.table.selection())["values"]
        for j in self.data_pelanggan:
            if j["nama"] == temp[1]:
                j["pembayaran"] = not j["pembayaran"]
                with open("data.json", "w") as k:
                    json.dump(self.data_pelanggan, k, indent=4)
                self.tampil()


class Menu(Frame):
    def __init__(self, root):
        super().__init__(root)
        self.temp = []
        self.root = root
        self.root.title("Kuy! Roti Bakar üçû")
        self.pelanggan = Pelanggan()
        self.kasir = Kasir()
        # self.root.config(background="light blue")

        self.btn_pelanggan = Button(
            self.root, text="Pelanggan", command=self.pelanggan.pelanggan
        )
        self.btn_pelanggan.grid(row=3, column=1, padx=10, pady=5)

        self.btn_kasir = Button(self.root, text="Kasir", command=self.kasir.kasir)
        self.btn_kasir.grid(row=3, column=2, padx=10, pady=5)

    


root = Tk()
# root.geometry("500x200")
menu = Menu(root)

menu.mainloop()
```
